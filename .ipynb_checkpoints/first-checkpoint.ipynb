```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.arange(1,15,1)
y=x**2
```

```python
plt.plot(x,y,color='r')
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=np.arange(1,15,1)
y1=x**2
y2=x**12

plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)

plt.show()
```

```python
#bar-plot

fruit={'apple':30,'manga':45, 'banana':10}
names=list(fruit.keys())
quantity=list(fruit.values())
```

```python
names,quantity
```

```python
plt.bar(names,quantity)
plt.show
```

```python
plt.bar(names,quantity, color = 'red')
plt.title('Distribution of fruits')
plt.show()
```

```python
#scatter-plot

x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,3,2,1,9]
b = [5,6,7,8,3,2,1,4,9]


plt.figure(figsize=(10,10)
plt.scatter(x,a,marker="4",s=500)
plt.scatter(x,b, s=200)
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.show()
```

```python
#histogram

data = [1,1,1,1,5,6,7,9,8,1,2,5,5,6,9,9,4,7,6,3]

plt.hist(data)
plt.show()
```

```python
#boxplot

one = [1,2,3,4,5,6,8,9]
two = [1,2,3,4,5,5,4,2,1]
three = [6,7,8,9,5,4,3,5]

data = list([one,two,three])
plt.boxplot(data)
plt.show()

#mean predicts
```

```python
#boxplot

one = [1,2,3,4,5,6,8,9]
two = [1,2,3,4,5,5,4,2,1]
three = [6,7,8,9,5,4,3,5]

data = list([one,two,three])
plt.violinplot(data, showmedians = True, showmeans = True)
plt.grid(True)
plt.title("Distribution of data")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()

#mean predicts
```

```python
#pie-chart

fruit = ['apple','manga','organe','grapes']
quality=[30, 45, 12, 100]
plt.pie(quality, labels=fruit, autopct='%0.1f%%',shadow=True,explode=(0,0.1,0,0.1))
plt.show()
```

```python
#pie-chart

fruit = ['apple','manga','organe','grapes']
quality=[30, 45, 12, 100]
pie1= plt.pie(quality, labels=fruit, autopct='%0.1f%%',radius=2)
pie2= plt.pie([5],colors='w',radius=1)

plt.show()
```

```python
#area plot

x = [10,20,30,40,50,60,70,80,90]
y = [1,1,2,5,6,5,7,5,6]

plt.stackplot(x,y)
plt.plot(x,y,color='green')
plt.show()
```
